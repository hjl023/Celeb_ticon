```python
import os
import numpy as np
import re
import json
import pandas as pd
from PIL import Image, ImageOps, UnidentifiedImageError
```

```python
filenamelist=[]
bbox_x_center_list=[]
bbox_y_center_list=[]
bbox_widthlist=[]
bbox_heightlist=[]
# 데이터프레임 딕셔너리 생성을 위한 리스트 

#파일 명을 가져오기 위해 resize 된 폴더에 접근
main_dir='.\\수정본'

emo_list=os.listdir(main_dir)

emo_list
for emo in emo_list: # 감성 리스트
    file_folder_path=os.path.join(main_dir,emo) # main_dir + 감성 
    print(emo)

    file_list=os.listdir(file_folder_path)

    #파일, 감정과 일치하는 라벨 데이터 접근
    label_dir=f'.\\라벨링데이터\EMOIMG_{emo}_SAMPLE\img_emotion_sample_data({emo}).json'

    f=open(label_dir)
    js = json.loads(f.read())
    df = pd.DataFrame(js) #데이터 프레임으로 열고

    for file in file_list: # 원본 bbox 좌표에 접근
        print(file)
        abs_size=df[df['filename']==file]['annot_A'].values[0]['boxes']
        print(abs_size)        
        # 파일 하나의 원본 bbox 좌표

        maxX=abs_size['maxX']
        maxY=abs_size['maxY']
        minX=abs_size['minX']
        minY=abs_size['minY']

        x_center = (maxX+minX)/2 # x y 중앙값과 w h 생성
        y_center = (maxY+minY)/2
        width = maxX-minX
        height = maxY-minY

        filenamelist.append(file)
        bbox_x_center_list.append(x_center)
        bbox_y_center_list.append(y_center)
        bbox_heightlist.append(height)
        bbox_widthlist.append(width) 



        
```

```python
dataf={}
dataf['file_name']=filenamelist
dataf['bbox_x_center_list']=bbox_x_center_list
dataf['bbox_y_center_list']=bbox_y_center_list
dataf['bbox_widthlist']=bbox_widthlist
dataf['bbox_heightlist']=bbox_heightlist

#데이터 프레임 생성
ds = pd.DataFrame(dataf)
ds.to_csv('./orizin_bbox_dot.csv',index=False)
```
